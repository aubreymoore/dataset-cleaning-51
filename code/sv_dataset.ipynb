```python
import supervision as sv
# from supervision.dataset.formats.yolo import YoloDataset
import cv2
```

```python
# Define the paths to your dataset
DATASET_DIR = '/home/aubrey/Desktop/Guam07-merged_results/YOLO-prepped/'
IMAGES_DIR = f"{DATASET_DIR}/images/val"
ANNOTATIONS_DIR = f"{DATASET_DIR}/labels/val"
DATA_YAML_PATH = f"{DATASET_DIR}/dataset.yaml"
```

```python
# Load the dataset using the `sv.DetectionDataset.from_yolo()` method
dataset = sv.DetectionDataset.from_yolo(
    images_directory_path=IMAGES_DIR,
    annotations_directory_path=ANNOTATIONS_DIR,
    data_yaml_path=DATA_YAML_PATH,
)
```

```python
# Now you can inspect and work with the loaded dataset.
print(f"Number of images in the dataset: {len(dataset)}")
print(f"Class names: {dataset.classes}")

# Access a single item from the dataset
image, labels, detections = dataset[0]
```

```python
# Visualize the loaded image and segmentation masks
image_bgr = cv2.imread(dataset.images[0])
mask_annotator = sv.MaskAnnotator()
label_annotator = sv.LabelAnnotator()

annotated_image = mask_annotator.annotate(
    scene=image_bgr.copy(),
    detections=detections
)
annotated_image = label_annotator.annotate(
    scene=annotated_image,
    detections=detections,
    labels=[dataset.classes[class_id] for class_id in detections.class_id]
)

# Display the image
sv.plot_image(annotated_image)
```
