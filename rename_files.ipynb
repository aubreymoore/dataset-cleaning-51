# rename_files.ipynb

Removes "IMG_" from all filenames within a YOLO dataset

```python
import glob
import os
```

```python
dir_path = '/home/aubrey/Desktop/crb-damage-detection-models/datasets/CRB005'
for path in glob.glob(f'{dir_path}/**/IMG_*', recursive=True):
    new_path = path.replace('IMG_', '')
    os.rename(path, new_path)
```
